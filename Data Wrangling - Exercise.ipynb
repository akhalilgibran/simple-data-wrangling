## Deal with Missing Data

How to deal with missing data?
1. drop data
   * drop the whole row
   * drop the whole column
2. Replace Data
   * replace it by mean
   * replace it by frequency
   * replace it based on other functions

#### Replace Data by Mean

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:",avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
df.head(5)
```

```python
#Calculate ABerage of Bore
avg_bore = df['bore'].astype("float").mean(axis=0)
print("Average of bore:",avg_bore)
```

```python
df['bore'].replace(np.nan, avg_bore, inplace=True)
df.head()
```

```python
#Calculate The Mean of Horsepower

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average of horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan,avg_horsepower,inplace=True)
df.head()
```

```python
#Calculate Peak RPM
avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis=0)
print("Average of peak rpm:",avg_peak_rpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peak_rpm,inplace=True)
df.head(10)
```

```python
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of Stroke:",avg_stroke)
```

```python
df['stroke'].replace(np.nan,avg_stroke,inplace=True)
df.head(10)
```

### Correct Data Format

```python
df.dtypes
```

As we can see above, some columns are not of the correct data type. Numerical variables should have type 'float' or 'int', and variables with strings such as categories should have type 'object'. For example, 'bore' and 'stroke' variables are numerical values that describe the engines, so we should expect them to be of the type 'float' or 'int'; however, they are shown as type 'object'. We have to convert data types into a proper format for each column using the "astype()" method.

#### Convert Data Types

```python
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["bore","stroke"]]=df[["bore","stroke"]].astype("float")
df[["price"]]=df[["price"]].astype("float")
df[["peak-rpm"]]=df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

## Data Normalization

Normalization is the process of transforming values of several variables into a similar range of values.

Example
To demonstrate normalization, let's say we want to scale the columns "length", "width", and "height"
<br> Target: would like to Normalize those variables so their value ranges from 0 to 1 </br>
<br>Approach: replace original value by (original value)/(maximum value) </br>

```python
df['length'] = df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
```

```python
df['height']=df['height']/df['height'].max()
```

```python
df[["length","width","height"]].head()
```

## Binning

Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.
<br>Example:</br>

<p>In our dataset, "horsepower" is a real valued variable ranging from 48 to 288, it has 57 unique values. What if we only care about the price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)? Can we rearrange them into three ‘bins' to simplify analysis?</p>

<br>We will use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins</br>

```python
df["horsepower"]=df["horsepower"].astype(int,copy=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.hist(df["horsepower"])

plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
bins=np.linspace(df['horsepower'].min(), df['horsepower'].max(),4)
bins
```

```python
#set group names
group_names=['Low','Medium','High']
```

```python
df['horsepower-binned']=pd.cut(df['horsepower'],bins,labels=group_names, include_lowest=True)
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline

plt.bar(group_names,df["horsepower-binned"].value_counts())

plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower bins")
```

```python
%matplotlib inline

plt.hist(df["horsepower"],bins=3)

plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

## Indicator Variable

<br>Indicator variable (or dummy variable) is a numerical variable used to label categories. They are called "dummies" because the numbers themselves don't have inherent meaning.</br>

<br>Why we use indicator variables?</br>

<br>So we can use categorical variables for regression analysis in the later modules.</br>

<br>Example:</br>
<br>We see column "fuel-type" has two unique values, "gas" or "diesel". Regression doesn't understand words, nly numbers. To use this attribute in regression analysis, we convert "fuel-type" into indicator variables.</br>

```python
df.columns
```

```python
dummy_variable_1=pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas','diesel':'fuel-type-diesel'},inplace=True)
dummy_variable_1.head()
```
